```python
import numpy as np
import pandas as pd
from scipy.linalg import solve
```

```python
def PowMax(A,x0,eps):
    x_1=A.dot(x0)
    xpast=np.copy(x0)
    itcount=1
    valpast=x_1[0]/xpast[0]
    xpast=np.copy(x_1)
    x_1=A.dot(x_1)
    val=x_1[0]/xpast[0]
    while itcount<999 and abs(valpast-val)>eps:
        valpast=np.copy(val)
        xpast=np.copy(x_1)
        x_1=A.dot(x_1)
        val=x_1[0]/xpast[0]
        itcount=itcount+1
    valpast=np.copy(val)
    xpast=np.copy(x_1)
    x_1=A.dot(x_1)
    val=x_1[0]/xpast[0]
    itcount=itcount+1
    
    return abs(val),itcount
```

```python
def ScaMax(A,x0,eps):
    x_1=A.dot(x0)
    xpast=np.copy(x0)
    AT=A.T
    itcount=1
    y_1=AT.dot(xpast)
    valpast=x_1.dot(y_1)/xpast.dot(x0)
    xpast=np.copy(x_1)
    x_1=A.dot(x_1)
    y_1=AT.dot(y_1)
    val=x_1.dot(y_1)/xpast.dot(y_1)
    while itcount<999 and abs(valpast-val)>eps:
        valpast=np.copy(val)
        xpast=np.copy(x_1)
        x_1=A.dot(x_1)
        y_1=AT.dot(y_1)
        val=x_1.dot(y_1)/xpast.dot(y_1)
        itcount=itcount+1
    valpast=np.copy(val)
    xpast=np.copy(x_1)
    x_1=A.dot(x_1)
    y_1=AT.dot(y_1)
    val=x_1.dot(y_1)/xpast.dot(y_1)
    itcount=itcount+1
    return abs(val), itcount
```

```python
#diagonal
number=4
a = np.array([[4,2,0,0],[2,5,2,0],[0,2,6,2],[0,0,2,7]])
x0=np.array([k+1 for k in range(number)])
a
```

```python
data = {'epsilon':[],'iterations PM': [],'||x-x~|| PM':[],'iterations SM': [],'||x-x~|| SM':[]}
df = pd.DataFrame(data)
eig=np.linalg.eig(a)

for number in range(2,7):
    eps=10**(-number)
    res=PowMax(a,x0,eps)
    rez=ScaMax(a,x0,eps)
    f=max(eig[0])-res[0]
    f1=max(eig[0])-rez[0]
    new_row={'epsilon':eps,'iterations PM': res[1],'||x-x~|| PM':abs(f),'iterations SM': rez[1],'||x-x~|| SM':abs(f1)}
    df=df.append(new_row,ignore_index=True)
df
```
